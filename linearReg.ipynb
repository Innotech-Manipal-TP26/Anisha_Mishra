```python
import numpy as np
class LinearRegression:
    def _init_(self):
        self.w = None
        self.b = 0.0

    def fit(self, X, y):
        X = np.asarray(X, dtype=float)
        y = np.asarray(y, dtype=float)
        if X.ndim == 1:
            X = X.reshape(-1, 1)

        Xb = np.c_[np.ones((X.shape[0], 1)), X]
        theta = np.linalg.lstsq(Xb, y, rcond=None)[0]

        self.b = float(theta[0])
        self.w = theta[1:]
        return self

    def predict(self, X):
        X = np.asarray(X, dtype=float)
        if X.ndim == 1:
            X = X.reshape(-1, 1)
        return X @ self.w + self.b
```

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/innotech taskphase 2/test.csv')
df=pd.read_csv('/content/drive/MyDrive/innotech taskphase 2/train.csv')
```

```python

features = ["GrLivArea", "OverallQual", "GarageCars", "TotalBsmtSF"]
X = df[features].to_numpy(dtype=float)
y = df["SalePrice"].to_numpy(dtype=float)


med = np.nanmedian(X, axis=0)
X = np.where(np.isnan(X), med, X)


rng = np.random.default_rng(0)
idx = rng.permutation(len(y))
split = int(0.8 * len(y))

train_idx, val_idx = idx[:split], idx[split:]
X_train, y_train = X[train_idx], y[train_idx]
X_val, y_val = X[val_idx], y[val_idx]


lr = LinearRegression().fit(X_train, y_train)
pred = lr.predict(X_val)
rmse = np.sqrt(np.mean((pred - y_val) ** 2))
print("RMSE:", rmse)
```
